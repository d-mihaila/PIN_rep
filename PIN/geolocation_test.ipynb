```python
import geopandas as gpd
from shapely.geometry import Point

def reverse_geocode_gadm(lat, lon, gpkg_path):
    """
    Given latitude & longitude, return:
    - smallest (most specific) admin region
    - its parent admin region
    - geometries (borders) for both
    """

    gdf = gpd.read_file(gpkg_path)

    # ensure lat/lon CRS
    if gdf.crs is None or gdf.crs.to_epsg() != 4326:
        gdf = gdf.to_crs(epsg=4326)

    point = Point(lon, lat)

    # Find the polygon containing the point
    matches = gdf[gdf.contains(point)]
    if matches.empty:
        return None

    row = matches.iloc[0]

    # Determine deepest admin level present
    levels = [
        (5, "NAME_5", "GID_5"),
        (4, "NAME_4", "GID_4"),
        (3, "NAME_3", "GID_3"),
        (2, "NAME_2", "GID_2"),
        (1, "NAME_1", "GID_1"),
        (0, "NAME_0", "GID_0"),
    ]

    smallest = None
    parent = None

    for i, (lvl, name_col, gid_col) in enumerate(levels):
        if name_col in row and row[name_col]:
            # smallest region
            smallest = {
                "level": lvl,
                "name": row[name_col],
                "gid": row[gid_col],
                "geometry": row.geometry
            }

            # find parent geometry in gdf using parent GID
            if i + 1 < len(levels):
                parent_lvl, parent_name_col, parent_gid_col = levels[i + 1]
                parent_gid = row[parent_gid_col]
                # lookup parent polygon by GID
                parent_row = gdf[gdf[parent_gid_col] == parent_gid]
                if not parent_row.empty:
                    parent = {
                        "level": parent_lvl,
                        "name": row[parent_name_col],
                        "gid": parent_gid,
                        "geometry": parent_row.iloc[0].geometry
                    }
            break

    return {
        "latitude": lat,
        "longitude": lon,
        "smallest_region": smallest,
        "parent_region": parent,
    }
```

```python
def plot_regions(result):
    import geopandas as gpd
    import matplotlib.pyplot as plt
    from shapely.geometry import Point
    from matplotlib.lines import Line2D
    from matplotlib.patches import Patch

    if not result:
        print("No region found.")
        return

    fig, ax = plt.subplots(figsize=(8, 8))

    # Parent region (draw FIRST)
    if result["parent_region"]:
        gpd.GeoSeries(
            [result["parent_region"]["geometry"]],
            crs="EPSG:4326"
        ).plot(
            ax=ax,
            facecolor="none",
            edgecolor="red",
            linewidth=1.5,
            zorder=1
        )

    # Smallest region (draw ON TOP)
    if result["smallest_region"]:
        gpd.GeoSeries(
            [result["smallest_region"]["geometry"]],
            crs="EPSG:4326"
        ).plot(
            ax=ax,
            facecolor="none",
            edgecolor="blue",
            linewidth=3,
            zorder=2
        )

    # Input point (draw LAST)
    ax.scatter(
        result["longitude"],
        result["latitude"],
        color="black",
        s=60,
        zorder=3
    )

    # Context image bounds from parent
    bounds = result["parent_region"]["geometry"].bounds
    ax.set_xlim(bounds[0], bounds[2])
    ax.set_ylim(bounds[1], bounds[3])

    # Custom legend
    legend_elements = [
        Patch(facecolor="none", edgecolor="blue", linewidth=3, label="Smallest region"),
        Patch(facecolor="none", edgecolor="red", linewidth=1.5, label="Parent region"),
        Line2D([0], [0], marker="o", color="black", linestyle="None", label="Input point")
    ]

    ax.legend(handles=legend_elements)
    ax.set_title("Administrative hierarchy and input coordinate")
    plt.show()
```

```python
plot_regions(result)
```

```python
def explore_regions(result):
    import geopandas as gpd
    from shapely.geometry import Point

    if not result:
        print("No region found.")
        return None

    geoms = []
    names = []
    levels = []

    # Parent region FIRST
    if result["parent_region"]:
        geoms.append(result["parent_region"]["geometry"])
        names.append(f"Parent (Level {result['parent_region']['level']}): "
                     f"{result['parent_region']['name']}")
        levels.append("parent")

    # Smallest region SECOND (on top)
    if result["smallest_region"]:
        geoms.append(result["smallest_region"]["geometry"])
        names.append(f"Smallest (Level {result['smallest_region']['level']}): "
                     f"{result['smallest_region']['name']}")
        levels.append("smallest")

    regions_gdf = gpd.GeoDataFrame(
        {
            "name": names,
            "level": levels
        },
        geometry=geoms,
        crs="EPSG:4326"
    )

    # Input point
    point_gdf = gpd.GeoDataFrame(
        {"name": ["Input coordinate"]},
        geometry=[Point(result["longitude"], result["latitude"])],
        crs="EPSG:4326"
    )

    # Plot regions
    m = regions_gdf.explore(
        column="level",
        categorical=True,
        categories=["parent", "smallest"],
        tooltip="name",
        style_kwds={"fillOpacity": 0.15, "weight": 3}
    )

    # Plot point
    point_gdf.explore(
        m=m,
        color="black",
        marker_kwds={"radius": 6},
        tooltip="name"
    )

    return m
```

```python
explore_regions(result)
```
